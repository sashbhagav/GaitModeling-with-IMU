```python
df_pressure = pd.read_csv('GP10_0.6_force.csv')
df_pressure = df_pressure[0:755]
```

```python
df_pressure = df_pressure['FP2_x']
```

```python
df_motion['Pressure'] = df_pressure
```

```python
df_motion.head()
```

```python
sns.set()
```

```python
sns.pairplot(df_motion)
```

```python
import warnings
warnings.filterwarnings('ignore')
from keras.models import Sequential
from keras.layers import Dense, Dropout
from sklearn.model_selection import train_test_split
import random 
```

```python
y = df_motion['Pressure']
X = df_motion.drop(columns=['Pressure'])
new_y = []
for i in y:
    if i < 0:
        new_y.append(abs(i))
    elif i == 0.0:
        new_y.append(random.uniform(8, 20))
    elif i > 20:
        new_y.append(i - 20)
    else:
        new_y.append(i)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, new_y, test_size=0.25, random_state=42)
```

```python
# validation = X_train.sample(frac=0.2, replace=False, random_state=1)
for i in X_train:
    _ = plt.scatter(df_motion[i], new_y)
#     v = plt.scatter(time, y)
    plt.title(i)
    plt.show()
```

```python
model = Sequential([
    (Dense(20, input_dim=X.shape[1], activation='relu')), 
    (Dense(10, activation='relu')), 
    (Dense(8, activation='relu')), 
    (Dense(2, activation='relu')),
    (Dense(1, activation='sigmoid'))
])
print(model.summary())
```

```python
model.compile(optimizer='sgd', loss='mean_squared_error',metrics=['accuracy'],)
model.fit(X_train, y_train, epochs=25, batch_size=25)
```

```python
def plot_data(title, times, df):   
    fig, axs = plt.subplots(3, figsize=(15,15))
    fig.suptitle('time vs. '+title)
#     plt.figure(figsize=(20,10))
    axs[0].plot(times, df[title+'(x)'], 'r')
    axs[1].plot(times, df[title+'(y)'], 'b')
    axs[2].plot(times, df[title+'(z)'], 'g')
    
    plt.xlabel('Time')
    plt.ylabel(title)
    plt.legend(['x dim', 'y dim', 'z dim'])
```

```python
plot_data('Acceleration',josh['Time(ms)'], josh)
```
